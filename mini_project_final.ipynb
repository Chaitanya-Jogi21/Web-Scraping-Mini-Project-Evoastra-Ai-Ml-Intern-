```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
web="https://ackodrive.com/collection/honda+cars/"
```

```python
r=requests.get(web)
```

```python
r.status_code
```

```python
scraper=BeautifulSoup(r.content, 'html.parser')
```

```python
cars=scraper.find_all('div',{"class":"BuyCarCard_card__AsGXF"})
```

```python
cars[1].find('a', {"class":"BuyCarCard_carName__SAJVh"}).get_text()
```

```python
len(cars)
```

```python
location=[]
for i in range(len(cars)):
    location.append(cars[i].find("p",{"class":"BuyCarCard_priceLabel__18XLX"}).get_text())
```

```python
for i in range(len(location)):
    location[i]=location[i][17:]
location
```

```python
car_name=[]
for i in range(len(cars)):  
    car_name.append(cars[i].find('a', {"class":"BuyCarCard_carName__SAJVh"}).get_text())
car_name
```

```python
car_price=[]
for i in range(len(cars)):  
    car_price.append(cars[i].find('p', {"class":"BuyCarCard_priceRange__2q8tm"}).get_text())
car_price
```

```python
car_name_lower=[]
for i in range(len(car_name)):
    car_name_lower.append(car_name[i].lower())
```

```python
car_name_lower
```

```python
car_price
```

```python
cars[2].find("div", {"class":"BuyCarCard_contentSpecifications__4ELYd"}).get_text()
```

```python
parent=scraper.find_all("div",{"class":"BuyCarCard_contentSpecifications__4ELYd"})
```

```python
fuel=[]
transmission=[]
for i in range(len(parent)):
    for p in parent[i].find_all("p", {"class":"BuyCarCard_specificationItem__yn5cu"}):
        text = p.get_text(strip=True)
        if "Petrol" in text or "Diesel" in text:
            fuel.append(text)
        elif "Manual" in text or "Automatic" in text:
            transmission.append(text)
```

```python
fuel
```

```python
transmission
```

```python
df=pd.DataFrame({"Car Name": car_name, "Price Range": car_price, "Fuel Type": fuel, "Transmission Type": transmission, "Location": location})
```

```python
df
```

```python
df['Year of Manufacture']=np.nan
df['Kilometers Travelled']=np.nan
df['Number of Owners']=np.nan
```

```python
df
```

```python
df.to_csv('Honda_Data')
```
